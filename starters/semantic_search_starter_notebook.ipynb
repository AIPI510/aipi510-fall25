```python
# Load an embedding model (you can use any embedding model)
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
# Provide your own dataset here. Decide on a chunking strategy and implement here.
texts = [
    "The Eiffel Tower is one of the most famous landmarks in Paris.",
    "Machine learning allows computers to learn from data.",
    "The chicken crossed the road to get to the other side."
]
```

```python
# Generate embeddings
embeddings = model.encode(texts)
```

```python
# Store embeddings in ChromaDB
import chromadb
from chromadb.config import Settings
client = chromadb.Client(Settings(anonymized_telemetry=False))
collection = client.create_collection("my_collection")
collection.add(documents=texts, embeddings=embeddings.tolist(), ids=[str(i) for i in range(len(texts))])
```

```python
# Define a search function
def semantic_search(query):
    query_embedding = model.encode([query])
    results = collection.query(query_embeddings=query_embedding.tolist(), n_results=1)
    return "\n\n".join(results['documents'][0])
```

```python
# Simple Gradio interface
import gradio as gr
demo = gr.Interface(
    fn=semantic_search,
    inputs=gr.Textbox(label="Enter your search query"),
    outputs=gr.Textbox(label="Top Matches"),
    title="Semantic Search Engine",
    description="Search over your custom dataset using semantic similarity."
)

demo.launch()
```
