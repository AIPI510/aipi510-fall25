## 1. Hypothesis Setting

Research Question:
- Drinking mushroom-based tea improves reaction time by 20ms on cognitive tests after just one week.
- Measure: reaction time in ms (continuous)

Null Hypothesis ($H_0$):
- There is no difference in reaction time between the group that drinks mushroom-based tea and the control group after one week.

Alternative Hypothesis ($H_1$):
- The group that drinks mushroom-based tea saw an improvement (decrease) in reaction time by 20ms after one week.

## 2. Choosing Statistical Test

This is an Independent samples t-test (one-sided)

Rationale:
- Comparing means of a continuous outcome (reaction time) between two independent groups (drink mushroom tea vs control)
- We specifically interested in improvement of reaction time, not just a difference
- A one-sided hypothesis is justified if there is no theoretical reason to expect worse reaction time due to the intervention.

Assumtions to check:
- Approximate normality of reaction time distribution in each group
- Similar variances between groups (can use Welch’s t-test if violated)
- Independence of observations (ensured by random assignment)

## 3. Literature Review for Effect Size

Use Cohen’s d for t-tests and simple group mean comparisons because we're comparing means between two groups and the outcomes are continuous.

Previous research suggests:
- The average human reaction time is 250ms (Cognitive psych studies (e.g., Posner 2020; Smith & Lee 2022))
- Standard deviations in reaction times typically range between 40-50ms

Using this, the standardized effect size (Cohen’s d) is estimated as:

d = Mean difference / Pooled standard deviation = 20/45 = 0.45 (medium effect size)

$$d = \frac{20}{45} \approx 0.45$$

## 4. Power Analysis

Determine the minimum sample size per group needed to detect an effect size of 0.45 with:
- Significance level (α): 0.05 (one-sided)
- Power (1 - β): 0.80

```python
from statsmodels.stats.power import TTestIndPower

# Parameters
effect_size = 0.45
alpha = 0.05
power = 0.80

# Power analysis
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='larger')

print(sample_size)
```

## Power Analysis Result

We need 62 participants per group (total of 124 participants) to detect a medium effect size with 80% power.
