Hypothesis for the study :

null  hypotheses: Turning off self-view doesn’t reduce stress more that guided meditation breaks

Alternate hypothesis : Turning off self-view  reduces stress more that guided meditation breaks

What statistical test will you perform?

Measure for the results : Perceived stress score: 0 to 40 

We will be using 
Cohen’s d	(Comparing means between two groups) 
 
Benchmark Values
0.2 = small
0.5 = medium
0.8 = large
