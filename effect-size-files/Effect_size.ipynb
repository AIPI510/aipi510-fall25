## Research Question:
Does turning off self-view using the ZoomZen plugin reduce perceived stress more than taking guided meditation breaks?

### Null Hypothesis (H₀):
There is no difference in Perceived Stress Scale (PSS) scores between participants who turn off self-view using the ZoomZen plugin and those who take guided meditation breaks.

H₀: The mean PSS after the self-view-off condition equals the mean PSS after the guided-meditation condition.


### Alternative Hypothesis (H₁):
Participants who turn off self-view using the ZoomZen plugin will have lower (better) PSS scores than participants who take guided meditation breaks.

H₁: The mean PSS after the self-view-off condition is significantly lower than after the guided-meditation condition.

## Statistical Test:
This is an Independent-Samples t-Test (one-sided)

## Assumptions :

1.Independence of observations : each participant’s PSS score is independent of others (no repeated measures or clustering).

2.Normality of PSS distribution (within each group): the PSS scores in each group are roughly normally distributed.

3.Homogeneity of variance (equal variances): the spread (variance) of PSS scores is similar across groups.

4.Measurement consistency: PSS is administered the same way for both groups — same scale version, timing, and context.

## Explore literature to find effect size : 

Prior research suggests:

    •    Viewing oneself on video increases self-evaluation and negative affect; hiding self-view is a primary recommendation to reduce Zoom-related fatigue/stress. (https://www.sciencedirect.com/science/article/pii/S2451958821000671?utm_source=chatgpt.com)


    •    Videoconferencing load/multitasking is positively associated with perceived stress. (https://pmc.ncbi.nlm.nih.gov/articles/PMC8645680/?utm_source=chatgpt.com)



    •    Mindfulness/meditation programs reduce PSS by ~2 points vs. controls in RCTs, roughly d ≈ 0.5 assuming SD ≈ 4; meta-analyses show small-to-moderate average effects with heterogeneity. (https://pmc.ncbi.nlm.nih.gov/articles/PMC10078965/?utm_source=chatgpt.com)


### Effect size estimate for our head-to-head trial:

Given 


(i) meditation’s typical small–moderate reduction in PSS and


(ii) the direct mechanism by which hiding self-view targets self-focused evaluation in Zoom meetings,


 
we conservatively plan for an incremental mean difference of 1.5–2.0 PSS points over meditation. 
With SD ≈ 4, this yields Cohen’s d ≈ 0.38–0.50. We therefore plan



**d = 0.40**


## Perform power analysis w/ effect size :

Determining the minimum sample size per group needed to detect an effect size of 0.4 with:
Significance level (α): 0.05 (one-sided)
Power (1 - β): 0.80


```python
from statsmodels.stats.power import TTestIndPower

# Parameters
effect_size = 0.4
alpha = 0.05
power = 0.80

# Power analysis
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='larger')
print(f"Required sample size per group: {sample_size:.2f}")
```

### Provide rationale for all decisions :

#### Rationale for picking the t-test as the statistical test

We are comparing the means of a continuous outcome (PSS score, 0–40) between two independent groups:

1.Participants who use the ZoomZen plugin to take guided meditation breaks

2.Participants who have self-view off during meetings

The research question specifically tests whether the self-view-off condition reduces perceived stress more (i.e., yields lower PSS scores), not just whether there is any difference.
A one-sided hypothesis is justified because there is no theoretical reason to expect turning off self-view to increase perceived stress.

#### Rationale for picking cohen's d

Cohen’s d was selected as the measure of effect size because it provides a standardized way to quantify the magnitude of the difference in Perceived Stress Scale (PSS) scores between groups.

It is particularly appropriate for t-test designs comparing two means (e.g., independent-samples or paired samples), allowing results to be interpreted and compared across studies regardless of the original PSS scale range (0–40).
