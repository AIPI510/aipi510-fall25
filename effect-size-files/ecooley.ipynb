### HydroHype Water Bottle Hypothesis testing
Claim: Carrying our smart water bottle increases average daily water consumption by at least 500 mL compared to standard bottles.

Meaure: Measure: mL of water consumed (continuous)

1. What are your null and alternative hypotheses?
- **Null hypothesis**: There is no difference in average daily water consumption between people who carry a HydroHype smart water bottle and people who carry a standard water bottle, or the difference is less than 500 mL.
- **Alternative hypothesis**: People who carry the HydroHype smart water bottle drink on average at least 500 mL more water than people who carry a standard water bottle.

2. What statistical test will you perform?
- A **t-test**, used to compare the means of two independent groups. 

3. Explore literature to find effect size

The article I found is [A Randomized Trial Evaluating the Use of a Smart Water Bottle to Increase Fluid Intake in Stone Formers](https://pubmed.ncbi.nlm.nih.gov/35283036/) on PubMed.

In this study, they had two groups, the group that used the smart bottle (41 participants) and the group that did not use the smart bottle, but only followed dietary recommendations (44 participants).

Although they ended up measuring total 24h urine volume, and not just whether water intake was increased using the smart bottle, we can still use this study to find an effect size for our HydroHype hypothesis test.

The mean increase is as follows:
- SB group: +1.37 L (95% CI: -0.51 to 3.25)
- DR group: +0.79 L (95% CI: -1.15 to 2.73)

To estimate the effect size, we can calculate Cohen's d using an assumed pooled standard deviation of 1.0 L (since the study doesn't mention a standard deviation)

```python
# Mean difference between groups
mean_diff = 1.37 - 0.79

# Calculate Cohen's d by dividing mean difference between groups with the pooled SD
cohen_d = mean_diff / 1.0

cohen_d
```

Cohen's d value came out to ~0.58, which is a medium to large effect.

4. Perform power analysis w/ effect size
- We will use the `TTestIndPower` package from the `statsmodels` python library. Excellent for power analysis of an independent samples t-test like this one. 
- We will plug in `alpha=0.05` and `power=0.8` as those were default values mentioned in the lecture slides. 

```python
from statsmodels.stats.power import TTestIndPower

alpha = 0.05
power = 0.8
alternative = 'larger'  # We use "larger" since this is a one-tailed test i.e. we are only testing if group A is greater than group B

ttest = TTestIndPower()

sample_size_per_group = ttest.solve_power(effect_size=cohen_d, alpha=alpha, power=power, alternative=alternative)

print(f"Sample size per group: {round(sample_size_per_group)}")
```

So, for our hypothesis test study with HydroHype, we will have a required sample size of 37 per group 74 total people as the entire sample size.
