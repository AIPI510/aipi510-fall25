# SnoozeSound Pillow — Power Analysis

**Claim:** Users who sleep with the embedded white-noise pillow fall asleep **15 minutes faster** on average than those without it.  

**Outcome measure:** Sleep onset latency (time-to-sleep) in minutes (continuous).

## Hypotheses

- **Null Hypothesis (H₀):** μ_treatment − μ_control = 0  (no difference in mean sleep onset latency)
- **Alternative Hypothesis (H₁):** μ_treatment − μ_control < 0  (treatment mean is smaller: people fall asleep faster with embedded white-noise pillow)

## Statistical test

We will use an independent two-sample t-test to compare average sleep onset latency between participants who use the white-noise pillow and those who use a standard pillow.
The dependent variable, time to fall asleep (minutes), is continuous, and the two groups are independent of each other.
Assuming the data are approximately normally distributed and have equal variances, the independent t-test is appropriate for determining whether the white-noise pillow group falls asleep significantly faster than the control group.

## Literature & Effect-size choice

From https://pmc.ncbi.nlm.nih.gov/articles/PMC5742584/, a broadband sound intervention in a within-subject crossover design (n = 18) produced an adjusted mean difference of –6.8 ± 2.7 minutes (p = 0.018) when controlling for order effects. Interpreting ±2.7 as the standard error and converting to an estimated SD of differences (≈11.46 minutes) yields Cohen’s 𝑑 = 6.8 / 11.46 ≈ 0.59. This suggests a medium-to-large effect.

## Assumptions for power calculation

Baseline mean SOL (control): 15.0 minutes — typical healthy adult value from sleep literature.

Expected treatment mean (white-noise pillow): 9.3 minutes — a 38% reduction from baseline (5.7 min difference) based on literature.

Assumed SD (each group): 10 minutes — plausible sleep-onset variability; we show sensitivity for other SDs below.

Pilot samples: n1 = n2 = 5 (small pilot assumption).

Hypothesis:treatment mean is smaller → one-sided test.

Alpha (one-sided): 0.05; Power: 0.80.

```python
from math import sqrt, ceil
from statsmodels.stats.power import TTestIndPower
import pandas as pd, numpy as np

n1, n2 = 5, 5
s1, s2 = 10**2, 10**2   # variances (SD=10)
u_control, u_treatment = 15.0, 9.3  # control mean, treatment mean

# pooled SD and Cohen's d
s = sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
d = (u_control - u_treatment) / s
print(f"Mean diff = {u_control - u_treatment:.2f} min, Cohen's d = {d:.3f}")

# power analysis
alpha = 0.05
power = 0.8
obj = TTestIndPower()

n = obj.solve_power(effect_size=d, alpha=alpha, power=power, ratio=1.0, alternative='larger')

print(f"One-sided: n per group = {ceil(n)} (total ≈ {ceil(n)*2})")
```
