# Study Power Analysis
## BrainBoost Gum Co.

**Claim**: Chewing our nootropic gum for 10 minutes before an exam improves memory recall scores more than chewing regular mint gum.

**Measure**: memory test score (continuous)

### What are your null and alternative hypotheses?
**H0 (Null)**: Chewing nootropic gun for 10 minutes before an exam does NOT improve memory recall scores than chewing regular mint gun.
> μ_BrainBoost – μ_mint = 0, where μ is the population mean

**H1 (Alternative)**: Chewing our nootropic gum for 10 minutes before an exam improves memory recall scores more than chewing regular mint gum.
> μ_BrainBoost – μ_mint > 0
