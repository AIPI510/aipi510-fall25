# Study Power Analysis
## BrainBoost Gum Co.

**Claim**: Chewing our nootropic gum for 10 minutes before an exam improves memory recall scores more than chewing regular mint gum.

**Measure**: memory test score (continuous)

### What are your null and alternative hypotheses?
**H0 (Null)**: Chewing nootropic gun for 10 minutes before an exam does NOT improve memory recall scores than chewing regular mint gun. 
> μ_BrainBoost – μ_mint <= 0, where μ is the population mean memory test score

**H1 (Alternative)**: Chewing our nootropic gum for 10 minutes before an exam improves memory recall scores more than chewing regular mint gum.
> μ_BrainBoost – μ_mint > 0


### What statistical test will you perform?

The comparison is between 2 *independent* groups (those that chewed nootropic vs regular mint) -> *between 2 sample subjects*

The memory test scores are *continuous*

assuming memory recall scores are normally distributed

independent, continuous, 2 groups ->
> Two sample T-test

### Explore literature to find effect size

Literature search on "nootripic gum effects on memory recall"

Found this one: https://pmc.ncbi.nlm.nih.gov/articles/PMC9133906/

> ...and effect size range of 0.4–0.6
> ... A medium effect size of 0.62 was estimated

So let's use the smaller side to be conservative, thus:
> d=0.4

### Perform power analysis w/ effect size

Use "standard" alpha for one-sided study
> alpha = 0.05

use Cohen's D = 0.4 from above literature ranges. See below for sensitivity if we were to adjust D.
> d = 0.4

Use "standard" power of 80%
> power = 0.8

```python
from statsmodels.stats.power import TTestIndPower
import math

power = TTestIndPower()

n_per_group = power.solve_power(effect_size=0.40, alpha=0.05, power=0.80)
n = math.ceil(n_per_group) * 2  # two groups, round up each group

print(f"Sample size per group: {math.ceil(n_per_group)}")
print(f"Total study participants needed: {n}")
```

```python
# sensitivity table for different effect sizes

import numpy as np, pandas as pd

d_vals = np.round(np.linspace(0.20, 0.60, 9), 2)
n_vals = [power.solve_power(effect_size=d, alpha=0.05, power=0.80)
          for d in d_vals]
sens = pd.DataFrame({'cohen_d': d_vals, 'n_per_group': np.ceil(n_vals).astype(int)})
display(sens)
```
