# PetPulse — Effect Size & Power Analysis

**Claim received:** Petting a dog for 5 minutes lowers heart rate more effectively than listening to a guided meditation.

**Measure:** Heart rate (continuous)

This notebook explicitly addresses the assignment checklist after receiving the claim:

- What are your null and alternative hypotheses?
- What statistical test will you perform?
- Explore literature to find effect size
- Perform power analysis with that effect size
- Provide rationale for all decisions

Reference used for effect size:

**Handlin, L., Hydbring-Sandberg, E., Nilsson, A., Ejdebäck, M., Jansson, A., & Uvnäs-Moberg, K. (2018).** The effects of a therapy dog on the blood pressure and heart rate of older residents in a nursing home. Anthrozoös, 31(4), 543–552.
Available at: https://publications.slu.se/?file=publ%2Fshow&id=96653

Handlin et al. (2018) found that therapy dog visits significantly reduced heart rate among older nursing-home residents, supporting a moderate physiological effect (Cohen’s d ≈ 0.5–0.6).


# Step 1: Hypotheses

**H0 (null):** There is no difference in average heart rate change between people who pet a dog and people who listen to meditation.

**H1 (alt):** People who pet a dog will have a bigger drop in heart rate.(H₁: μ_dog − μ_meditation > 0).
Handlin reports clear heart-rate decreases after dog visits thus directional hypothesis.

# Step 2: Choose a test

Comparing two groups (dog vs meditation) and our outcome (heart rate) is a number.

The right test is an **independent samples t-test**.

Using a one-sided test because the claim says “more effectively,” meaning it is expected that one group is lower.


# Step 3: Effect size from research

Handlin et al. (2018) reported significant decrease in heart rate over time in the dog group, but it did not present Cohen's d. I assume an effect size of about 0.6 for heart rate reduction when people interacted with a dog, because the paper clearly provides proven results. 


```python
effect_size_d = 0.6
```


# Step 4: Power analysis

Finding how many people are needed in each group to detect this effect.

```python
import numpy as np
import pandas as pd
from statsmodels.stats.power import TTestIndPower


alpha = 0.05  # significance level
power = 0.8   # desired power

power_analysis = TTestIndPower()

n = power_analysis.solve_power(effect_size=effect_size_d, power=power, alpha=alpha)
num_per_group = int(np.ceil(n))

print(f"Effect size (Cohen's d): {effect_size_d}")
print(f"Alpha: {alpha}, Power: {power}")
print(f"Estimated participants per group: {num_per_group}")
print(f"Total participants needed: {num_per_group*2}")
```
