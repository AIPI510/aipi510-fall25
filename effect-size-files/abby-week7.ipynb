# 1. What are your null and alternative hypotheses?
- **H0:** The mean reduction in heart rate after 5 minutes of petting a dog equals the mean reduction after listening to a guided meditation.
- **HA:** The mean reduction in heart rate after 5 minutes of petting a dog is greater than the mean reduction after listening to a guided meditation.


# 2. What statistical test will you perform?
I chose a paired-samples t-test because each participant will meet both conditions (meditating and petting a dog). The test evaluates the mean of heart-rate differences, accounting per-person for the correlation between repeated measures and aligns with the hypothesis that dog-petting yields a larger reduction.


# 3. Explore literature to find effect size
Per the publication "Brain and heart activity during interactions with pet dogs: A portable electroencephalogram and heart rate variability study" at https://www.sciencedirect.com/science/article/pii/S0167876024001168, heart rate rises modestly (~4–6 bpm) typically corresponding to medium effects (Cohen’s d≈ 0.4–0.5). So I will use medium effect at 0.5.

```python
# 4. Perform power analysis w/ effect size
from statsmodels.stats.power import TTestPower
import numpy as np

effect_size = 0.5  # Cohen's d for paired differences from literature
alpha = 0.05       # one-sided test to match directional hypothesis
power = 0.80       # conventional target

analysis = TTestPower()  # paired design reduces to one-sample test on difference scores
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='larger')
required_participants = int(np.ceil(sample_size))
print(f'Required sample size (per paired condition): {sample_size:.2f} -> recruit at least {required_participants} participants')
```

# 5. Provide rationale for all decisions
PetPulse expects dog petting to lower heart rate more than meditation, so I framed a hypothesis and matched it with a one-sided paired t-test on per-person differences. Literature on therapy dog sessions versus quiet controls clusters around Cohen's d ≈ 0.5, so I planned for that medium effect. Using α = 0.05 and 80% power keeps Type I errors low while targeting a realistic sample size, giving the 23-participant recruitment goal from the power analysis.
