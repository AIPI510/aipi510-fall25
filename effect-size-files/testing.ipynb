1. What are your null and alternative hypothesis?

Null Hypothesis (H₀): There is no difference in time to sleep between those who you the embedded white nopise pillow and those who do not.
\
Alternative Hypothesis (H₁): There is a difference betweenn those who use the embedded white pillow and those whop do not in terms of time to sleep.

2. What statistical test will you perform?
This is a Independent samples t-test (one-sided)

Rationale:
 - Comparing means of a continuous outcome (time to sleep) between two independent groups (treatment vs. control).
 - While the team is specifically interested in improvement, the sleep of the users could get worse or the pillow could be uncomfortable and the sleep score could be worse.
 - A two-sided hypothesis is justified because users might think the pillow is uncomfortable or might now like white noise and so it might take them longer to fall asleep.


To determine effect size:

d = Mean difference / Pooled standard deviation = 21 / 28 = .75 (medium effect)

We're going to be a little more conservative and go with 0.5.

```python
from statsmodels.stats.power import TTestIndPower

# Parameters
effect_size = 0.5
alpha = 0.05
power = 0.80

# Power analysis
analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, alternative='two-sided')
```
