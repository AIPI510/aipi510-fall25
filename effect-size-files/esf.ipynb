## Effect Size Analysis for MoodMate App
Null hypothesis (H_0) where p_new = p_current
Alternartive hypothesis (H_1) where p_new = 0.75*p_current

I have chosen p_current to be 0.083 which follows the NIMH statistics of 8.3% of adults having a major depressive episode in 2021. Therefore, for the alternative hypothesis to be true, p_new = 0.75 * 0.083 = 0.06225.

Effect size is calculated with these values.

## Why NIMH
The NIMH is a trusted government agency with access to US citizens medical data, so it is a valuable and useful source of data.

NIMH: https://www.nimh.nih.gov/health/statistics/major-depression

## Statistical Testing Method
A z-test is ideal here because the goal is to compare a mean of a control dataset to a experimental test group's mean. Z-tests are great for large data sets and comparing the mean of a population to a single test group. Z tests use normal distribution for comparison.


```python
# Initial Imports
from math import sqrt
from statsmodels.stats.proportion import proportion_effectsize
from statsmodels.stats.power import TTestIndPower
```

## Defining Variables
Probabilities are defined using logic and NIMH data explained above. Alpha and power are set to 0.05 and 0.8 respectively which are widely accepted for early stage analysis.

```python
# Defining probabilities
p_new = 0.06225
p_current = 0.083

# Standard values from lectures and slides
alpha = 0.05
power = 0.8
```

## Finding Effect Size
Using statsmodels.stats.proportion.proportion_effectsize which takes two proportions and method, I find Cohen's Recommendation effect size by setting method = 'normal'.

```python
# Calculating Effect Size
d = proportion_effectsize(p_new, p_current, method='normal')
print(f'Effect size: {d}')
```

## Finding Required Sample Size
I am using TTestIndPower() as was provided by the example in the slides. Effect size is calculated above using Cohen's Recommendation. I am using Cohen's Recommendation because it is best for early stage anaylsis to find minimum sample size which is the objective here.

```python
# Calculating Required Sample Size
obj = TTestIndPower()
n = obj.solve_power(effect_size=d, alpha=alpha, power=power, 
                    ratio=1, alternative='two-sided')

print('Sample size/Number needed in each group: {:.3f}'.format(n))
```
